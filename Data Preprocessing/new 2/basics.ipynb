THE MAPPING FROM TEXTUAL DATA TO REAL VALUED VECTORS IS CALLED FEATURE EXTRACTION

TF-IDF VECTORIZER
1.TERM FREQUENCY(TF)-(NUMBER OF TIMES TERM APPEARS IN A DOCUMENT)/(NUMBER OF TERMS N A DOCUMENT)
2.INVERSE DOCUMENT FREQUECY(IDF) - LOG(N/n) - WHERE N NUMBER OF DOCUMENTS, n NUMBER OF DOCUMENTS A TERM T HAS APPEARED IN
3.THE IDF VALUE OF A RARE WORD IS HIGH, WHEREAS THE IDE OF A FREAQUENT WORD IS LOW
