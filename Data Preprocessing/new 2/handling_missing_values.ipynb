MISSING VALUES - OCCOUR IN WHEN NO DATA VALUE IS STORED FOR A VARIABLE(FEATURE) IN AN OBSERVATION

```python
import pandas as pd
import numpy as np
```

```python
# These are the headers which you need to link to your dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(r"E:\Programming\Machine Learning\Data Preprocessing\new 2\car_dataset.csv", names=headers)
df.head()
```

```python
# replace ? to nan
df.replace("?", np.nan, inplace=True)
df.head()
```

```python
df.to_csv('car_dataset_header.csv', index=False)
```

```python
missing_data = df.isnull()
missing_data.head()
```

TO DEAL MISSING VALUES
1. DROP - THE WHOLE ROW, COLUMN
2. REPLACE - WITH AVERAGE, MEAN, MODE

```python
df.dtypes
```

```python
avg_nor_loss = df['normalized-losses'].astype("float").mean()
avg_nor_loss
```

```python
df['normalized-losses'].replace(np.nan, avg_nor_loss, inplace=True)
df.head()
```

```python
avg_bore = df['stroke'].astype('float').mean(axis=0)
avg_bore
```

```python
df['stroke'].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis = 0)
avg_horsepower
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['body-style'].value_counts()
```

```python
# to find more number of doors
df['num-of-doors'].value_counts().idxmax()
```

```python
# replace the missing doors by values most frequent
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
df.dropna(subset=['price'], inplace=True, axis=0)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

HANDLING MISSING VALUES USING SKLEARN

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']])
df[['normalized-losses','stroke','bore','horsepower','peak-rpm']] = imputer.transform(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']])
```

```python
df.head()
```

```python
imputer_most_freq = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df[['num-of-doors']] = imputer_most_freq.fit_transform(df[['num-of-doors']])
```

```python
# third row row having price value missing
df.dropna(subset=['price'], axis=0, inplace=True)
```

```python
# reset the index because we dropped four rows
df.reset_index(drop=True, inplace=True)
```
