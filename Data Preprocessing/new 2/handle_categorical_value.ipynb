```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv(r"E:\Programming\Machine Learning\Data Preprocessing\new 2\Salary_Dataset.csv")
dataset.head()
```

ONE-HOT ENCODING

```python
# get dummy variables
country_dummy = pd.get_dummies(dataset['country'])
country_dummy
```

```python
# concatnate the dataframes
dataset_new = pd.concat([dataset, country_dummy], axis = 1)
dataset_new.head()
```

```python
dataset_new.drop('country',axis=1,inplace=True)
dataset_new.head()
```

```python
dataset_new = dataset_new[['Australia', 'Canada', 'Dubai', 'USA', 'Salary', 'YearsExperience', 'Purchased']]
dataset_new.head()
```

USING SCIKIT LEARN

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dataset.head()
```

```python
le = LabelEncoder()
dataset['Purchased'] = le.fit_transform(dataset['Purchased'])
```

ONE-HOT ENCODING ON COUNTRY COLUMN

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
dataset = (ct.fit_transform(dataset)).astype("O")
dataset
```

```python
type(dataset)
```

```python
# changing the dataset to pandas datadframe
dataset = pd.DataFrame(ct.fit_transform(dataset))
dataset
```
