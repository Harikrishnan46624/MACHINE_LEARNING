```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
brest_cancer = pd.read_csv(r"E:\Programming\Machine Learning\Data Preprocessing\exploratry data anlysis.csv")
brest_cancer.head()
```

```python
# removing the unnamed column
brest_cancer.drop(columns='Unnamed: 32', axis = 1, inplace=True)
```

```python
brest_cancer.info()
```

```python
brest_cancer.drop(columns='id', axis=1, inplace=True)
```

```python
brest_cancer.isnull().sum()
```

```python
#take summery of Descriptive Statistics
brest_cancer.describe()
```

```python
brest_cancer['diagnosis'].value_counts()
```

```python
label_encode = LabelEncoder()

lables = label_encode.fit_transform(brest_cancer['diagnosis'])

brest_cancer['target'] = lables

brest_cancer.drop(columns='diagnosis', axis=1, inplace=True)
```

Outliers Detection

Box plot for visualizing the outliers in the dataset

```python
for column in brest_cancer:
    plt.figure()
    brest_cancer.boxplot([column])
```

As we can see here that most of the skewed features have Outliers

Correlation Matrix

```python
correlation_matrix = brest_cancer.corr()
```

```python
# constructing a heat map to visualize the correlation matrix
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, cbar=True, fmt='.1f', annot=True, cmap='Blues')
plt.savefig('correlatio heat map')
```


We can remove the features if they have high +ve or -ve correlation between them
