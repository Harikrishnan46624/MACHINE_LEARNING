DATA VISULIZATION LIBRARY

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

VISUALIZE DATA

```python
# setting a theme for plots
sns.set_theme()
```

```python
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time', hue= 'smoker', style = 'smoker', size = 'size')
```

SCATTER PLOT

```python
#house predict dataset
from sklearn.datasets import fetch_california_housing
```

```python
house_cali = fetch_california_housing()
print(house_cali)
```

```python
house = pd.DataFrame(house_cali.data, columns=house_cali.feature_names)
house['price'] = house_cali.target
```

```python
house.head()
```

DISTRIBUTION PLOT

```python
sns.displot(house['price'])
```

```python
sns.distplot(house['price'])
```

CORRELATION:
POSTIVE CORRELATION,
NEGATIVE CORRELATION

HEAT MAP

```python
correlation = house.corr()
```

```python
# constructing a heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
