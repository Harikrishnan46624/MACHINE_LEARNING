```python
import pandas as pd
```

```python
df = pd.read_csv("E:\Programming\Machine Learning\Libaries\pandas\weather_data.csv")
df
```

FILL NONE VALUES

```python
#fill none values with zero
new_df = df.fillna(0)
new_df
```

```python
#fill none values with zero specified
new_df = df.fillna({
    'temperature' : 0,
    'windspeed': 0,
    'event' : 'no event'
})
new_df
```

```python
#fill none values with upper data , forward fill
new_df = df.fillna(method='ffill')
new_df
```

```python
#fill none values with down data , backward fill
new_df = df.fillna(method='bfill')
new_df
```

```python
#fill none values with right data 
new_df = df.fillna(method='bfill', axis='columns')
new_df
```

```python
#fill none values with right data 
new_df = df.fillna(method='ffill', axis='columns')
new_df
```

```python
#fill none values with left data 
new_df = df.fillna(method='ffill', axis='columns')
new_df
```

```python
#fill none values with down data with limit, backward fill
new_df = df.fillna(method='ffill', limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
