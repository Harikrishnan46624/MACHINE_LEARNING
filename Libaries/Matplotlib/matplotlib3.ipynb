```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
blood_sugar = np.array([113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 123])
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95)
```

```python
exp_vals = [1400, 600, 300,  410, 250]
exp_labels = ['Home Rent', "Food", 'Phone', 'Car', 'Others']
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0,0.5,0.5,0])
plt.show()
```

SAVING THE PIE CHART

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.0f%%',explode=[0,0.5,0,0,0])
plt.savefig("pie_study.png", bbox_inches="tight", pad_inches=2, transparent=True)
```
