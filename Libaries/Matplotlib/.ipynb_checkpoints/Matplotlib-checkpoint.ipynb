```python
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

PLOTTING DATA

```python
# sine wave
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

```python
#adding title x-axis and y-axis labels
plt.plot(x, y)
plt.xlabel("angle")
plt.ylabel("sine wave")
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

BAR PLOT

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
langauge = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70] 
ax.bar(langauge, people)
plt.xlabel("Languages")
plt.ylabel("Number of people")
plt.show()
```

PIE CHART

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langauge = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=langauge, autopct='%1.1f%%')
plt.show()
```

SCATTER PLOT

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x, y, color = 'g')
ax.scatter(x, z, color = 'b')
plt.show()
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap = 'Blues')
plt.show()
```
