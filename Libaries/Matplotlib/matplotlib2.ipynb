```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49, 46]
```

```python
plt.xlabel('Day')
plt.ylabel('temprature')
plt.title('Whethar')
plt.plot(x, y)
```

```python
plt.xlabel('Day')
plt.ylabel('temprature')
plt.title('Whethar')
plt.plot(x, y, color='green', linewidth=5, linestyle='dotted')
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('temprature')
plt.title('Whethar')
plt.plot(days, max_t,)
plt.plot(days, min_t,)
plt.plot(days, avg_t,)
```

```python
plt.xlabel('Days')
plt.ylabel('temprature')
plt.title('Whethar')
plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Average")
plt.legend(loc = "best", fontsize="small")
plt.grid()
```

```python
import numpy as np
```

```python
company = ['Goog', 'Amzn', 'Mstf', 'Fb']
revenue = [90, 136, 89, 27]
profit = [40, 2, 34, 12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US tech stocks")      
plt.bar(ypos, revenue, label="revenue") 
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US tech stocks")      
plt.bar(ypos, revenue, label="revenue") 
plt.bar(ypos, profit, label ="profit")
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US tech stocks")      
plt.bar(ypos-0.2, revenue,width=0.4, label="revenue") 
plt.bar(ypos+0.2, profit,width=0.4,label ="profit")
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US tech stocks")      
plt.barh(ypos-0.2, revenue, label="revenue") 
plt.barh(ypos+0.2, profit,label ="profit")
plt.legend()
```
