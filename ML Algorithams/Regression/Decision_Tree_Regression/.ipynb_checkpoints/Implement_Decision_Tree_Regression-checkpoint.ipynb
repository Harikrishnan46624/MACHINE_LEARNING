```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"E:\Programming\Machine Learning\ML Algorithams\Regression\Decision_Tree_Regression\5_Position_Salaries.csv")
```

```python
df.head()
```

```python
X = df.iloc[:, 1:2].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor()
```

```python
reg.fit(X, y)
```

```python
reg.predict([[6.5]])
```

Visualising the model (lower resolution (wrong))

```python
plt.scatter(X, y, color='red')
plt.plot(X, reg.predict(X), color='blue')
plt.title('Salary vs Level (Decision Tree)')
plt.ylabel('Salary')
plt.xlabel('Level')
```

Visualising the model (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, reg.predict(X_grid), color='blue')
plt.title('Salary vs Level (Decision Tree)')
plt.ylabel('Salary')
plt.xlabel('Level')
```

```python
from sklearn import tree
fig, ax = plt.subplots(figsize=(15, 10))
tree.plot_tree(reg, filled=True, feature_names=df.columns)
plt.show()
```
