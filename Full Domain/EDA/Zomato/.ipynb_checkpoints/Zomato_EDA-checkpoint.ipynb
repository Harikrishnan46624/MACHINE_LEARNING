IMPORTING ANOTHER DATASET COUNTRY

```python
df_country = pd.read_excel(r"E:\Programming\Machine Learning\Full Domain\EDA\Country-Code.xlsx")
df_country.head()
```

COMBINING TWO DATASET

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df
```

```python
##To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

which country has maximimum orders plot by pie chart

```python
## Pie Chart
plt.pie(country_val,labels=country_names ,autopct='%1.2f%%')
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color', data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
