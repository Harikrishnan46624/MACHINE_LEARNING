Partitioning Clustering or centroid-based method

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

Generate sample data for clustering

```python
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

Create and fit the K-Means clustering model

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

Predict cluster labels

```python
labels = kmeans.labels_
```

 Calculate silhouette score (a clustering evaluation metric)

```python
silhouette_avg = silhouette_score(X, labels)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='X', c='red')
plt.title(f'K-Means Clustering\nSilhouette Score: {silhouette_avg:.2f}')
plt.show()
```
